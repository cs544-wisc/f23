```python
from google.cloud import bigquery
```

```python
%load_ext google.cloud.bigquery
```

```python
bq = bigquery.Client()
```

```python
dry = bigquery.QueryJobConfig(dry_run=True)
```

```python
q = bq.query("""
SELECT *
FROM bigquery-public-data.geo_us_boundaries.counties
""", job_config=dry)
q.total_bytes_processed / 1024**2 # MB
```

```python
q = bq.query("""
SELECT county_geom
FROM bigquery-public-data.geo_us_boundaries.counties
""", job_config=dry)
q.total_bytes_processed / 1024**2 # MB
```

```python
q = bq.query("""
SELECT county_name
FROM bigquery-public-data.geo_us_boundaries.counties
""", job_config=dry)
q.total_bytes_processed / 1024**2 # MB
```

```python
cost_cap = bigquery.QueryJobConfig(maximum_bytes_billed=100*1024**2) # 100 MB
q = bq.query("""
SELECT *
FROM bigquery-public-data.geo_us_boundaries.counties
LIMIT 10
""", job_config=cost_cap)
q.to_dataframe()
```

```python
cost_cap = bigquery.QueryJobConfig(maximum_bytes_billed=200*1024**2) # 200 MB
q = bq.query("""
SELECT *
FROM bigquery-public-data.geo_us_boundaries.counties
LIMIT 10
""", job_config=cost_cap)
q.to_dataframe()
```

```python
# 55 is FIPS for WI
q = bq.query("""
SELECT *
FROM bigquery-public-data.geo_us_roads.us_national_roads
WHERE state_fips_code = 55
""", job_config=dry)
q.total_bytes_processed / 1024**2 # MB
```
